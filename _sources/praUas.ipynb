### **Deteksi Missing Value**

Pada tahap data preprocessing penulis akan mengidentifikasi type data pada kolom yang ada pada dataset terlebih dahulu. Identifikasi ini diperlukan agar program yang nanti dijalankan tidak salah mengganti data missing valuenya. Setelahnya penulis merubah dataset yang awalnya merupakan missing value menjadi ada, agar nanti ketika sudah diproses dan dijadikan model muncul hasil yang lebih baik. berikut ini adalah program yang akan mengganti missing valuenya.

```python
# Daftar kolom numerik berdasarkan informasi dari kamu
numeric_cols = [
    'Age', 'Number of sexual partners', 'First sexual intercourse', 'Num of pregnancies',
    'Hormonal Contraceptives (years)', 'IUD (years)', 'STDs (number)',
    'STDs: Number of diagnosis', 'STDs: Time since first diagnosis', 'STDs: Time since last diagnosis'
]

# Kolom target yang tidak akan diimputasi
target_cols = ['Biopsy', 'Hinselmann', 'Schiller', 'Cytology']

# Sisanya kita anggap biner (berdasarkan bool di daftar kamu)
binary_cols = [col for col in df_copy.columns if col not in numeric_cols + target_cols]

# Imputasi numerik: isi dengan mean
for col in numeric_cols:
    mean_val = df_copy[col].mean()
    df_copy[col].fillna(mean_val, inplace=True)

# Imputasi biner: isi dengan modus
for col in binary_cols:
    mode_val = df_copy[col].mode()[0]
    df_copy[col].fillna(mode_val, inplace=True)

# Cek hasilnya
print("Total missing value setelah imputasi:", df_copy.isnull().sum().sum())
```

```python
missing_per_column_copy = df_copy.isnull().sum()

missing_per_column_copy = missing_per_column_copy[missing_per_column_copy > 0]

print(missing_per_column_copy)
```

Bisa dilihat setelah menjalankan program yang sudah dibuat missing value yang awalnya banyak tadi menjadi sudah kosong.

## **Model**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
from imblearn.over_sampling import SMOTE
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

### *SMOTE(Synthetic Minority Over-sampling Technique)*

Pada pembuatan model Decision Tree saya menggunakan metode SMOTE dikarenakan data yang saya miliki dominan pada sisi tertentu saja. Metode Smote ini juga akan saya lakukan bukan saat proses Data Preprocessing, karena jika dilakukan pada saat proses tersebut, akan terjadi kebocoran data yang akan menyebabkan rusaknya data yang ingin diproses. Proses SMOTE hanya dilakukan pada saat training Model nya saja tanpa menyentuh data aslinya.

### **Model Algoritma Decision Tree**

Berikut ini program untuk pembuatan model Algoritma Decision Tree

```python
decision_tree_X = X.copy()
decision_tree_y = y.copy()
```

#### *Tanpa SMOTE*

```python
X_train, X_test, y_train, y_test = train_test_split(
    decision_tree_X, decision_tree_y, test_size=0.2, random_state=42, stratify=y
)

# 4. Buat model Decision Tree
clf = DecisionTreeClassifier(random_state=42, max_depth=3)  # max_depth bisa disesuaikan
clf.fit(X_train, y_train)

# 5. Prediksi
y_pred = clf.predict(X_test)

# 6. Evaluasi hasil
print("Akurasi:", accuracy_score(y_test, y_pred))
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

#### *SMOTE*

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(decision_tree_X, decision_tree_y)

# 4. Split data SMOTE ke train/test
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

# 5. Buat model dan latih
clf = DecisionTreeClassifier(max_depth=5, class_weight='balanced', random_state=42)
clf.fit(X_train, y_train)

# 6. Prediksi
y_pred = clf.predict(X_test)

# 7. Evaluasi performa
print("Akurasi:", accuracy_score(y_test, y_pred))
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

### **Model Algoritma Naive Bayes**

```python
naive_bayes_X = X.copy()
naive_bayes_y = y.copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    naive_bayes_X, naive_bayes_y, stratify=y, test_size=0.2, random_state=42
)

# 4. Terapkan SMOTE ke training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# 5. Buat model Naive Bayes dan latih
nb_model = GaussianNB()
nb_model.fit(X_train_smote, y_train_smote)

# 6. Prediksi terhadap test set
y_pred = nb_model.predict(X_test)

# 7. Evaluasi hasil model
print("Akurasi:", accuracy_score(y_test, y_pred))
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

### **Model Algoritma KNN(K-Nearest Neighbors)**

```python
knn_X = X.copy()
knn_y = y.copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    knn_X, knn_y, stratify=y, test_size=0.2, random_state=42
)

# 3. Terapkan SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# 4. Latih model KNN tanpa scaling
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_smote, y_train_smote)

# 5. Prediksi dan evaluasi
y_pred = knn.predict(X_test)

print("Akurasi:", accuracy_score(y_test, y_pred))
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
