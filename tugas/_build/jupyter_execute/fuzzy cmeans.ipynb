# ***Fuzzy C-mean***

## **Pengertiann Fuzzy C-Means**

**Fuzzy C-Means** (FCM) adalah metode clustering (pengelompokan data) yang termasuk dalam kategori soft clustering. Berbeda dengan metode clustering tradisional seperti K-Means yang menempatkan setiap titik data ke dalam satu dan hanya satu cluster, FCM memungkinkan setiap titik data memiliki derajat keanggotaan pada cluster yang berbeda-beda. Derajat keanggotaan ini biasanya bernilai antara 0 dan 1, di mana nilai yang mendekati 1 menunjukkan tingkat keanggotaan yang tinggi pada cluster tersebut.

## **Langkah Langkah Algoritma Fuzzy C-Means**

Konsep dasar dari algortima ini menemukan nilai derajat keanggotaan dari semua data yang ada pada centroid/Nilai tengah, serta mencari titik centroid dari setiap cluster yang sudah ditentukan. Langkah ini bersifata iteratif(berulang) dengan langkah-langkah berikut:

1. Inisialisasi
    - Menentukan jumlah cluster yang akan digunakan
    - Menentukan nilai derajat anggota pada setiap centroid cluster. 
2. Menghitung Centroid:
    - Titik centroid sendiri diperoleh dari perhitungan nilai derajat anggota, nanti penulis sertakan rumusnya
3. Memperbarui Matriks Keanggotaan:
    - Setelah perhitungan centroid sudah dilakukan, maka perbarui nilai derajat keanggotaan pada setiap anggota dengan menggunakan hasil Centroid tadi. Nilai dari derajat keanggotaan yang semakin dekat, maka akan semakin tinggi nilainya. 
4. Evaluasi: 
    - Biasanya pengulangan perhitungan tersebut terus dilakukan sampai dengan batas iterasi yang ditentukan atau sampai dengan mencapai nilai error yang telah mencapai nilai minimum. Nilai error ini biasanay ditentukan dari perbandingan antar nilai derajat keanggotaan, atau nilai centroid dari suatu cluster serta bisa juga gabungan antara keduanya yaitu fungsi objektif. Langkah ini biasa disebut dengan perhitungan nilai konvergensi

## **Rumus Rumus Fuzzy-Cmeans**

### *Rumus Perhitungan Centroid*


Setelah matriks keanggotaan ($U$) diinisialisasi atau dihitung, pusat *cluster* ($v_i$) diperbarui menggunakan rumus berikut:

$$v_i = \frac{\sum_{k=1}^{n} (u_{ik})^m x_k}{\sum_{k=1}^{n} (u_{ik})^m}$$

Di mana:
* $v_i$: Pusat *cluster* ke-$i$.
* $n$: Jumlah total titik data.
* $u_{jk}$: Derajat keanggotaan titik data ke-$k$ terhadap klaster $j$.
* $m$: Parameter *fuzziness* (dalam kasus ini, $m=2$).
* $x_k$: Titik data ke-$k$.


* $\sum_{k=1}^{n} (\mu_{ik})^m x_k$: Penjumlahan dari setiap titik data $x_k$ yang dibobot oleh derajat keanggotaannya terhadap *cluster* $i$ yang dipangkatkan $m$.
* $\sum_{k=1}^{n} (\mu_{ik})^m$: Jumlah bobot keanggotaan dari semua titik data terhadap *cluster* $i$.



Rumus ini pada dasarnya menghitung rata-rata tertimbang (weighted average) dari semua titik data untuk menentukan posisi centroid, di mana bobotnya adalah derajat keanggotaan.

### *Rumus Perhitungan Nilai Derajat Keanggotaan*

Setelah pusat *cluster* ($V$) dihitung, matriks keanggotaan ($U$) diperbarui menggunakan rumus berikut:

$$\mu_{ik} = \frac{1}{\sum_{j=1}^{c} \left( \frac{\|x_k - v_i\|}{\|x_k - v_j\|} \right)^{\frac{2}{m-1}}}$$

Di mana:
* $\mu_{ik}$: Derajat keanggotaan titik data $x_k$ pada *cluster* $i$.
* $\|x_k - v_i\|$: Jarak Euclidean antara $x_k$ dan pusat *cluster* $v_i$.
* $\|x_k - v_j\|$: Jarak Euclidean antara $x_k$ dan pusat *cluster* $v_j$ (untuk setiap *cluster* $j$).

Rumus ini menunjukkan bahwa derajat keanggotaan titik data $x_k$ pada suatu *cluster* $i$ berbanding terbalik dengan jaraknya dari pusat *cluster* $v_i$, dan berbanding terbalik dengan jaraknya dari semua pusat *cluster* lainnya. Semakin dekat $x_k$ ke $v_i$, semakin tinggi $\mu_{ik}$.


### *Rumus Perhitungan Fungsi Objektif*

Fungsi objektif yang diminimalkan oleh FCM adalah:

$$J_m(U, V) = \sum_{k=1}^{n} \sum_{i=1}^{c} (\mu_{ik})^m \|x_k - v_i\|^2$$

Di mana:
* $J_m(U, V)$: Nilai fungsi objektif.
* $(\mu_{ik})^m$: Derajat keanggotaan $x_k$ pada *cluster* $i$ yang dipangkatkan dengan $m$. Ini memberikan bobot yang lebih besar pada titik data yang memiliki keanggotaan kuat pada suatu *cluster*.
* $\|x_k - v_i\|^2$: Kuadrat jarak Euclidean antara titik data $x_k$ dan pusat *cluster* $v_i$.

Tujuan algoritma ini adalah mencari $U$ dan $V$ yang meminimalkan $J_m$.

* Hitung Pembilang (Numerator): $\sum_{k=1}^{N} (u_{1k})^m \cdot x_k$

    Ini adalah perkalian element-wise antara $(u_{1k})^m$ dan $x_k$, kemudian dijumlahkan.
$$
\begin{aligned}
\text{Numerator} &= (u_{11}^2 \cdot x_1) + (u_{12}^2 \cdot x_2) + (u_{13}^2 \cdot x_3) + (u_{14}^2 \cdot x_4) + (u_{15}^2 \cdot x_5) \\

\end{aligned}
$$

    Ini adalah penjumlahan semua nilai $(u_{1k})^m$.
$$
\begin{aligned}

&= 0.25 + 0.49 + 0.64 + 0.49 + 0.36 \\

\end{aligned}
$$

$$
\begin{aligned}

## **Implementasi Fuzzy C-means**

### *Penggunaan Algortima Fuzzy C-Means Pada Contoh data*

Berikut ini adalah implementasi Penggunaan algoritma Fuzzy C-Means Clustering dengan menggunakan python. Data yang digunakan pada code dibawah ini sama dengan data yang sudah penulis hitung sebelumnya. Perbedaannya hanya pada penggunaan bahasa python untuk mendapatkan hasil yang cepat dan juga akurat. Penulis juga menggunakan fungsi objektif serta menghitung konvergensinya agar tidak melakukan iterasi sampai dengan ambang batas maksimal. Sedangkan untuk ambang batas maksimalnya pada code dibawah ini adalah 100 iterasi, akan tetapi karena sudah mencapai batas konvergensinya maka hanya dilakukan 8 iterasi saja.

```python
import numpy as np

# -------------------------------------------
# DATA & PARAMETER
# -------------------------------------------
data = np.array([
    [1, 2],
    [2, 3],
    [3, 4],
    [6, 7],
    [7, 8]
])

# Inisialisasi keanggotaan awal (baris: cluster, kolom: data)
membership = np.array([
    [0.4345, 0.3963, 0.2582, 0.5758, 0.5531],   # cluster 1
    [0.5655, 0.6037, 0.7418, 0.4242, 0.4469]    # cluster 2
])

m = 2
max_iter = 100
epsilon = 1e-5  # ambang konvergensi


# -------------------------------------------
# FUNGSI UTAMA
# -------------------------------------------
def compute_centroids(data, membership, m):
    um = membership ** m
    return (um @ data) / um.sum(axis=1, keepdims=True)


def update_membership(data, centroids, m):
    n_samples = data.shape[0]
    n_clusters = centroids.shape[0]
    expo = 2.0 / (m - 1.0)
    dist = np.linalg.norm(data[:, None, :] - centroids[None, :, :], axis=2)
    dist = np.where(dist == 0, 1e-10, dist)
    membership = np.empty((n_clusters, n_samples))
    for j in range(n_clusters):
        ratio = (dist[:, j, None] / dist) ** expo
        membership[j, :] = 1.0 / ratio.sum(axis=1)
    return membership


# -------------------------------------------
# ITERASI FCM SAMPAI KONVERGEN
# -------------------------------------------
for iteration in range(max_iter):
    prev_centroids = compute_centroids(data, membership, m)
    membership = update_membership(data, prev_centroids, m)
    new_centroids = compute_centroids(data, membership, m)

    diff = np.linalg.norm(new_centroids - prev_centroids)
    print(f"\nIterasi {iteration+1}")
    for idx, c in enumerate(new_centroids, 1):
        print(f"  Centroid Cluster {idx}: {c}")
    print(f"  Derajat keanggotaan:\n{np.round(membership, 4)}")
    print(f"  Selisih konvergensi: {diff:.6f}")

    if diff < epsilon:
        print(f"\nKonvergen tercapai pada iterasi {iteration+1} dengan selisih {diff:.6f}")
        break
else:
    print("\nMaksimal iterasi tercapai tanpa konvergen.")

final_centroids = new_centroids
final_membership = membership
final_diff = diff

```

```python
def initialize_membership(n_samples, n_clusters):
    u = np.random.dirichlet(np.ones(n_clusters), size=n_samples)
    return u.T  # bentuk: (n_clusters, n_samples)

def compute_centroids(data, u, m):
    um = u ** m
    return (um @ data) / um.sum(axis=1)[:, None]

def update_membership(data, centroids, m):
    dist = np.linalg.norm(data[:, None, :] - centroids[None, :, :], axis=2) + 1e-10
    exp = 2 / (m - 1)
    temp = (dist[:, :, None] / dist[:, None, :]) ** exp
    u_new = 1 / temp.sum(axis=2)
    return u_new.T
```

```python
def fuzzy_c_means(data, n_clusters=3, m=2, max_iter=100, epsilon=1e-5):
    n_samples = data.shape[0]
    u = initialize_membership(n_samples, n_clusters)
    
    for i in range(max_iter):
        centroids = compute_centroids(data, u, m)
        u_new = update_membership(data, centroids, m)

        if np.linalg.norm(u_new - u) < epsilon:
            print(f"Konvergen pada iterasi ke-{i+1}")
            break
        u = u_new

        print(f"Iterasi {i+1} - Centroid:")
        for idx, c in enumerate(centroids):
            print(f"  Cluster {idx+1}: {c}")

    return centroids, u
```
